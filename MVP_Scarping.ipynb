**import the mvp tables from 1992**

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

URL = "https://www.basketball-reference.com/awards/awards_{}.html"
years = list(range(1992,2023))

for year in years:
  url = URL.format(year)
  data = requests.get(url)

  with open("/content/mvp/{}.html".format(year),'w+') as f:
    f.write(data.text)
```

```python
import IPython
IPython.display.HTML(filename='/content/team/1992.html')
```

```python
dfs =[]
years = list(range(1992,2023))
for year in years:
  with open("/content/mvp/{}.html".format(year)) as f:
    page = f.read()
  soup = BeautifulSoup(page,"html.parser")
  soup.find('tr',class_="over_header").decompose()
  mvp_table = soup.find(id="mvp")
  mvp = pd.read_html(str(mvp_table))[0]
  mvp["Year"] = year

  dfs.append(mvp)

```

```python
mvps = pd.concat(dfs)
mvps.head()
```

```python
mvps.to_csv("mvps.csv")
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
player_stats_url = "https://www.basketball-reference.com/leagues/NBA_{}_per_game.html"
years = list(range(1992,2023))

for year in years:
  url = player_stats_url.format(year)
  data = requests.get(url)
  with open("/content/player/{}.html".format(year), "w+") as f:
        f.write(data.text)
```

```python
dfs =[]
years = list(range(1992,2023))
for year in years:
  with open("/content/player/{}.html".format(year)) as f:
    page = f.read()
  soup = BeautifulSoup(page,"html.parser")
  soup.find('tr',class_="thead").decompose()
  player_table = soup.find(id="per_game_stats")
  player = pd.read_html(str(player_table))[0]
  player["Year"] = year

  dfs.append(player)
```

```python
players = pd.concat(dfs)
```

```python
players.tail()
```

```python
players.to_csv("players.csv")
```

```python
team_stats_url = "https://www.basketball-reference.com/leagues/NBA_{}_standings.html"
for year in years:
    url = team_stats_url.format(year)

    data = requests.get(url)

    with open("/content/team/{}.html".format(year), "w+") as f:
        f.write(data.text)
dfs = []
for year in years:
    with open("team/{}.html".format(year)) as f:
        page = f.read()

    soup = BeautifulSoup(page, 'html.parser')
    soup.find('tr', class_="thead").decompose()
    e_table = soup.find_all(id="divs_standings_E")[0]
    e_df = pd.read_html(str(e_table))[0]
    e_df["Year"] = year
    e_df["Team"] = e_df["Eastern Conference"]
    del e_df["Eastern Conference"]
    dfs.append(e_df)

    w_table = soup.find_all(id="divs_standings_W")[0]
    w_df = pd.read_html(str(w_table))[0]
    w_df["Year"] = year
    w_df["Team"] = w_df["Western Conference"]
    del w_df["Western Conference"]
    dfs.append(w_df)
teams = pd.concat(dfs)
teams.tail()
```
